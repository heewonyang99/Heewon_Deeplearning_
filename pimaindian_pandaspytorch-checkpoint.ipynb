```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch.nn.init as init

import numpy

from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

import tensorflow as tf

numpy.random.seed(3)
tf.random.set_seed(3)


dataset = numpy.loadtxt(r'C:\Users\data set/pima-indians-diabetes.csv',delimiter = ",", dtype = numpy.float32)

X = torch.FloatTensor(dataset[:,0:-1])
Y = torch.FloatTensor(dataset[:,[-1]])


model = nn.Sequential(
    nn.Linear(8,12), 
    nn.ReLU(), 
    nn.Linear(12,8), 
    nn.ReLU(), 
    nn.Linear(8,1), 
    nn.Sigmoid())

epochs = 100

loss_fnc = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr = 0.02)

for epoch in range (epochs+ 1):
    optimizer.zero_grad()
    hypothesis = model(X)
    loss = loss_fnc(hypothesis, Y)
    loss.backward()
    optimizer.step()
    
    print('Epoch {:4d}/{}   Cost: {:.6f}'.format(epoch, epochs, loss.item()))
```
